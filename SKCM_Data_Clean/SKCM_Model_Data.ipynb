## 得到 PRAD_Model_TF_Expression_Features.csv

```python
# Author: Weilin He
# Description: Generate Model_TF_Expression_Features.csv by replacing HGNC_Symbol with TF_Symbol.

import pandas as pd

base_path = "/Users/heweilin/Desktop/M-28-SKCM//NTU_DATA_CLEANED/"

# Load the PRAD expression CSV file
prad_file = base_path + "SKCM_TOIL_RSEM_TPM_Levels_After_DEG_DMA.csv"
prad_df = pd.read_csv(prad_file, dtype=str)

# Load the TF_Target CSV file
tf_target_file = base_path + "TF_Target.csv"
tf_target_df = pd.read_csv(tf_target_file, dtype=str)

# Merge dataframes on HGNC_Symbol to replace with TF_Symbol
merged_df = pd.merge(prad_df, tf_target_df[['HGNC_Symbol', 'TF_Symbol']], on='HGNC_Symbol', how='inner')

# Replace HGNC_Symbol with TF_Symbol and drop original HGNC_Symbol
merged_df.drop('HGNC_Symbol', axis=1, inplace=True)

# Reorder columns to place TF_Symbol first
cols = merged_df.columns.tolist()
cols.insert(0, cols.pop(cols.index('TF_Symbol')))
merged_df = merged_df[cols]

# Save the resulting DataFrame to a new CSV file
output_file = base_path + "SKCM_Model_TF_Expression_Features.csv"
merged_df.to_csv(output_file, index=False)

# Print the first five rows to check the result
print(merged_df.head())
```

## 换成——

## 把后缀为11的数据列全部删掉，然后删掉剩下数据中数据的-01后缀即可

```python
import pandas as pd

# Step 1: Load the CSV file
file_path = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_Model_Methylation_Features.csv"
df = pd.read_csv(file_path)

# Step 2: First replace '-' with '_' in all column names (this is crucial before further suffix processing)
df.columns = [col.replace("-", "_") for col in df.columns]

# Step 3: Remove columns that end with '_11'
df = df[[col for col in df.columns if not col.endswith("_11")]]

# Step 4: Remove '_01' suffix from remaining sample columns (not from 'Probe_ID')
new_columns = []
for col in df.columns:
    if col.endswith("_01"):
        new_columns.append(col[:-3])  # remove '_01'
    else:
        new_columns.append(col)
df.columns = new_columns

# Step 4: Remove '_06' suffix from remaining sample columns (not from 'Probe_ID')
new_columns = []
for col in df.columns:
    if col.endswith("_06"):
        new_columns.append(col[:-3])  # remove '_06'
    else:
        new_columns.append(col)
df.columns = new_columns

# Step 5: Display first 5 rows
print("Modified DataFrame (first 5 rows):")
print(df.head())

# Step 6: Overwrite the original CSV file
df.to_csv(file_path, index=False)
```
