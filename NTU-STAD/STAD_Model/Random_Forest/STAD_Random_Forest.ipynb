```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import VarianceThreshold, f_classif, SelectKBest
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import warnings

# 忽略警告
warnings.filterwarnings('ignore')

# 设置路径
input_dir = '/Users/heweilin/Desktop/M-28-STAD/NTU_DATA_CLEANED/'
output_dir = '/Users/heweilin/Desktop/M-28-STAD/STAD_Model/Random_Forest/'

# 确保输出目录存在
os.makedirs(output_dir, exist_ok=True)

# 加载数据
print("加载数据...")
# MIR100HG表达水平数据
mir_exp = pd.read_csv(os.path.join(input_dir, 'STAD_Model_MIR100HG_Expression_Levels.csv'))
# 基因表达数据
gene_exp = pd.read_csv(os.path.join(input_dir, 'STAD_Model_Gene_Expression_Features.csv'))
# 甲基化数据
methylation = pd.read_csv(os.path.join(input_dir, 'STAD_Model_Methylation_Features.csv'))

print("MIR100HG表达数据形状:", mir_exp.shape)
print("基因表达数据形状:", gene_exp.shape)
print("甲基化数据形状:", methylation.shape)

# 数据预处理
print("\n数据预处理...")

# 提取标签
y_data = mir_exp[['Sample_ID', 'Group']]
y_data['Group'] = y_data['Group'].map({'High': 1, 'Low': 0})

# 处理基因表达数据
print("\n处理基因表达数据...")
gene_exp_pivot = gene_exp.copy()

# 检查是否存在MIR100HG基因，并将其从特征中删除
if 'MIR100HG' in gene_exp_pivot.index:
    print("检测到MIR100HG基因，将其从特征中删除...")
    gene_exp_pivot = gene_exp_pivot.drop('MIR100HG', axis=0)

gene_exp_pivot = gene_exp_pivot.set_index('HGNC_Symbol')
gene_exp_pivot = gene_exp_pivot.transpose()
gene_exp_pivot = gene_exp_pivot.reset_index()
gene_exp_pivot = gene_exp_pivot.rename(columns={'index': 'Sample_ID'})

# 处理甲基化数据
print("\n处理甲基化数据...")
methylation_pivot = methylation.copy()
methylation_pivot = methylation_pivot.set_index('Probe_ID')
methylation_pivot = methylation_pivot.transpose()
methylation_pivot = methylation_pivot.reset_index()
methylation_pivot = methylation_pivot.rename(columns={'index': 'Sample_ID'})

# 合并数据
print("\n合并数据...")
# 先合并MIR和基因表达
tmp_merged = y_data.merge(gene_exp_pivot, on='Sample_ID', how='inner')
print(f"MIR和基因表达合并后的形状: {tmp_merged.shape}")

# 然后合并甲基化数据
merged_data = tmp_merged.merge(methylation_pivot, on='Sample_ID', how='inner')
print(f"加入甲基化数据后的形状: {merged_data.shape}")

# 检查合并后的数据
if merged_data.shape[0] == 0:
    print("警告: 合并后的数据为空! 检查Sample_ID是否匹配。")
    import sys
    sys.exit(1)

print(f"合并后的数据形状: {merged_data.shape}")
print(f"样本数量: {merged_data.shape[0]}")
print(f"特征数量 (包括Sample_ID和Group): {merged_data.shape[1]}")

# 分割数据为X和y
X = merged_data.drop(['Sample_ID', 'Group'], axis=1)
y = merged_data['Group']
sample_ids = merged_data['Sample_ID']

# 为特征添加前缀
gene_features = gene_exp_pivot.columns.tolist()[1:]  # 跳过Sample_ID
methylation_features = methylation_pivot.columns.tolist()[1:]  # 跳过Sample_ID

gene_prefix = {feature: f"Gene_{feature}" for feature in gene_features}
methylation_prefix = {feature: f"Methylation_{feature}" for feature in methylation_features}

# 重命名列
X = X.rename(columns={**gene_prefix, **methylation_prefix})

print(f"\n添加前缀后的特征数量: {X.shape[1]}")
print(f"基因表达特征数量: {len(gene_features)}")
print(f"甲基化特征数量: {len(methylation_features)}")

# 确保MIR100HG相关特征被删除
mir100hg_cols = [col for col in X.columns if 'MIR100HG' in col]
if len(mir100hg_cols) > 0:
    print(f"删除以下MIR100HG相关特征: {mir100hg_cols}")
    X = X.drop(mir100hg_cols, axis=1)

# 获取基因表达和甲基化特征的列
gene_columns = [col for col in X.columns if col.startswith('Gene_')]
methylation_columns = [col for col in X.columns if col.startswith('Methylation_')]

# 划分训练集和测试集
print("\n划分训练集和测试集...")
X_train, X_test, y_train, y_test, ids_train, ids_test = train_test_split(
    X, y, sample_ids, test_size=0.2, random_state=42, stratify=y
)

print(f"训练集样本数: {X_train.shape[0]}")
print(f"测试集样本数: {X_test.shape[0]}")

# 特征选择函数 - 分别为基因表达和甲基化数据进行特征选择
def select_features(X_train, y_train, var_threshold=0.005, k_gene=200, k_meth=100):
    # 分离基因表达和甲基化特征
    X_train_gene = X_train[gene_columns]
    X_train_meth = X_train[methylation_columns]
    
    print("特征选择前:")
    print(f"总特征数量: {X_train.shape[1]}")
    print(f"基因表达特征数量: {X_train_gene.shape[1]}")
    print(f"甲基化特征数量: {X_train_meth.shape[1]}")
    
    # 1. 对基因表达特征进行方差过滤
    selector_var_gene = VarianceThreshold(threshold=var_threshold)
    X_train_gene_var = selector_var_gene.fit_transform(X_train_gene)
    gene_var_features = X_train_gene.columns[selector_var_gene.get_support()].tolist()
    print(f"方差过滤后的基因表达特征数量: {len(gene_var_features)}")
    
    # 2. 对甲基化特征进行方差过滤
    selector_var_meth = VarianceThreshold(threshold=var_threshold)
    X_train_meth_var = selector_var_meth.fit_transform(X_train_meth)
    meth_var_features = X_train_meth.columns[selector_var_meth.get_support()].tolist()
    print(f"方差过滤后的甲基化特征数量: {len(meth_var_features)}")
    
    # 3. 对基因表达特征进行ANOVA F检验
    X_train_gene_var_df = pd.DataFrame(X_train_gene_var, columns=gene_var_features, index=X_train.index)
    k_gene_actual = min(k_gene, X_train_gene_var_df.shape[1])
    selector_f_gene = SelectKBest(f_classif, k=k_gene_actual)
    _ = selector_f_gene.fit_transform(X_train_gene_var_df, y_train)
    gene_selected_features = X_train_gene_var_df.columns[selector_f_gene.get_support()].tolist()
    print(f"ANOVA选择后的基因表达特征数量: {len(gene_selected_features)}")
    
    # 4. 对甲基化特征进行ANOVA F检验
    X_train_meth_var_df = pd.DataFrame(X_train_meth_var, columns=meth_var_features, index=X_train.index)
    k_meth_actual = min(k_meth, X_train_meth_var_df.shape[1])
    selector_f_meth = SelectKBest(f_classif, k=k_meth_actual)
    _ = selector_f_meth.fit_transform(X_train_meth_var_df, y_train)
    meth_selected_features = X_train_meth_var_df.columns[selector_f_meth.get_support()].tolist()
    print(f"ANOVA选择后的甲基化特征数量: {len(meth_selected_features)}")
    
    # 合并所选特征
    all_selected_features = gene_selected_features + meth_selected_features
    print(f"合并后的所选特征总数: {len(all_selected_features)}")
    print(f"基因表达特征: {len(gene_selected_features)}")
    print(f"甲基化特征: {len(meth_selected_features)}")
    
    # 创建选择矩阵
    combined_support = np.zeros(X_train.shape[1], dtype=bool)
    for i, feature in enumerate(X_train.columns):
        if feature in all_selected_features:
            combined_support[i] = True
    
    return combined_support, all_selected_features

# 使用训练集进行特征选择
print("\n在训练集上进行特征选择...")
feature_support, selected_features = select_features(X_train, y_train)

# 应用特征选择
X_train_selected = X_train.loc[:, feature_support]
X_test_selected = X_test.loc[:, feature_support]

print(f"选择后的特征数量: {X_train_selected.shape[1]}")
print("选择的基因表达特征数量:", len([f for f in X_train_selected.columns if f.startswith('Gene_')]))
print("选择的甲基化特征数量:", len([f for f in X_train_selected.columns if f.startswith('Methylation_')]))

# 特征重要性分析
def feature_importance_analysis(model, X, feature_names):
    # 获取特征重要性
    importances = model.feature_importances_
    
    # 创建一个DataFrame来存储特征重要性
    feature_importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances
    })
    
    # 按重要性排序
    feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)
    
    # 计算百分比
    feature_importance_df['Percentage'] = feature_importance_df['Importance'] / feature_importance_df['Importance'].sum() * 100
    
    return feature_importance_df

# 训练随机森林模型
print("\n训练随机森林模型...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train_selected, y_train)

# 分析特征重要性
print("\n特征重要性分析...")
feature_importance = feature_importance_analysis(rf_model, X_train_selected, X_train_selected.columns)

# 分类特征并计算每类特征的总重要性
gene_importance = feature_importance[feature_importance['Feature'].str.startswith('Gene_')]
methylation_importance = feature_importance[feature_importance['Feature'].str.startswith('Methylation_')]

gene_importance_sum = gene_importance['Importance'].sum()
methylation_importance_sum = methylation_importance['Importance'].sum()
total_importance = gene_importance_sum + methylation_importance_sum

# 打印特征重要性统计
print("\n特征重要性百分比:")
print(f"基因表达特征: {gene_importance_sum / total_importance * 100:.2f}%")
print(f"甲基化特征: {methylation_importance_sum / total_importance * 100:.2f}%")

# 打印按类别分组的Top20特征
print("\n基因表达Top20特征:")
print(gene_importance.head(20).to_string(index=False))

print("\n甲基化Top20特征:")
print(methylation_importance.head(20).to_string(index=False))

print("\n总体Top20特征:")
print(feature_importance.head(20).to_string(index=False))

# 保存特征重要性结果
gene_importance.to_csv(os.path.join(output_dir, 'STAD_Random_Forest_Gene_Importance.csv'), index=False)
methylation_importance.to_csv(os.path.join(output_dir, 'STAD_Random_Forest_Methylation_Importance.csv'), index=False)
feature_importance.to_csv(os.path.join(output_dir, 'STAD_Random_Forest_All_Features_Importance.csv'), index=False)

# 在测试集上评估模型
print("\n在测试集上评估模型...")
y_pred = rf_model.predict(X_test_selected)
y_prob = rf_model.predict_proba(X_test_selected)[:, 1]

# 计算指标
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)
cm = confusion_matrix(y_test, y_pred)

# 打印评估结果
print("\n测试集性能指标:")
print(f"准确率 (Accuracy): {accuracy:.4f}")
print(f"精确率 (Precision): {precision:.4f}")
print(f"召回率 (Recall): {recall:.4f}")
print(f"F1分数: {f1:.4f}")
print(f"AUC: {auc:.4f}")
print("混淆矩阵:")
print(cm)

# 保存模型性能指标
with open(os.path.join(output_dir, 'SKCM_RandomForest_Model_Performance.txt'), 'w') as f:
    f.write(f"测试集性能指标:\n")
    f.write(f"准确率 (Accuracy): {accuracy:.4f}\n")
    f.write(f"精确率 (Precision): {precision:.4f}\n")
    f.write(f"召回率 (Recall): {recall:.4f}\n")
    f.write(f"F1分数: {f1:.4f}\n")
    f.write(f"AUC: {auc:.4f}\n")
    f.write("混淆矩阵:\n")
    f.write(f"{cm[0][0]} {cm[0][1]}\n")
    f.write(f"{cm[1][0]} {cm[1][1]}\n")

# 交叉验证
print("\n执行交叉验证...")
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results = {
    'fold': [],
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1': [],
    'auc': []
}

fold = 1
for train_idx, val_idx in skf.split(X, y):
    print(f"\n执行第{fold}折交叉验证...")
    
    # 获取当前fold的训练和验证数据
    X_fold_train, X_fold_val = X.iloc[train_idx], X.iloc[val_idx]
    y_fold_train, y_fold_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # 在fold内进行特征选择
    print(f"Fold {fold} - 特征选择...")
    fold_feature_support, fold_selected_features = select_features(X_fold_train, y_fold_train)
    
    # 应用特征选择
    X_fold_train_selected = X_fold_train.loc[:, fold_feature_support]
    X_fold_val_selected = X_fold_val.loc[:, fold_feature_support]
    
    # 训练模型
    print(f"Fold {fold} - 训练模型...")
    fold_model = RandomForestClassifier(
        n_estimators=100,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        random_state=42,
        n_jobs=-1
    )
    fold_model.fit(X_fold_train_selected, y_fold_train)
    
    # 预测
    y_fold_pred = fold_model.predict(X_fold_val_selected)
    y_fold_prob = fold_model.predict_proba(X_fold_val_selected)[:, 1]
    
    # 计算性能指标
    fold_accuracy = accuracy_score(y_fold_val, y_fold_pred)
    fold_precision = precision_score(y_fold_val, y_fold_pred)
    fold_recall = recall_score(y_fold_val, y_fold_pred)
    fold_f1 = f1_score(y_fold_val, y_fold_pred)
    fold_auc = roc_auc_score(y_fold_val, y_fold_prob)
    
    # 打印本折结果
    print(f"Fold {fold} - 性能指标:")
    print(f"  准确率: {fold_accuracy:.4f}")
    print(f"  精确率: {fold_precision:.4f}")
    print(f"  召回率: {fold_recall:.4f}")
    print(f"  F1分数: {fold_f1:.4f}")
    print(f"  AUC: {fold_auc:.4f}")
    
    # 保存结果
    cv_results['fold'].append(fold)
    cv_results['accuracy'].append(fold_accuracy)
    cv_results['precision'].append(fold_precision)
    cv_results['recall'].append(fold_recall)
    cv_results['f1'].append(fold_f1)
    cv_results['auc'].append(fold_auc)
    
    # 分析特征重要性
    fold_importance = feature_importance_analysis(fold_model, X_fold_train_selected, X_fold_train_selected.columns)
    
    # 保存本折的特征重要性
    fold_importance.to_csv(os.path.join(output_dir, f'PRAD_RandomForest_Fold{fold}_Feature_Importance.csv'), index=False)
    
    fold += 1

# 计算交叉验证的平均性能
cv_results_df = pd.DataFrame(cv_results)
cv_mean = cv_results_df.mean()

print("\n交叉验证平均性能:")
print(f"平均准确率: {cv_mean['accuracy']:.4f}")
print(f"平均精确率: {cv_mean['precision']:.4f}")
print(f"平均召回率: {cv_mean['recall']:.4f}")
print(f"平均F1分数: {cv_mean['f1']:.4f}")
print(f"平均AUC: {cv_mean['auc']:.4f}")

# 保存交叉验证结果
cv_results_df.to_csv(os.path.join(output_dir, 'STAD_Random_Forest_CV_Results.csv'), index=False)

print("\n分析完成！所有结果已保存到:", output_dir)
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import VarianceThreshold, f_classif, SelectKBest
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import warnings

# 忽略警告
warnings.filterwarnings('ignore')

# 设置路径
input_dir = '/Users/heweilin/Desktop/M-28-STAD/NTU_DATA_CLEANED/'

# 确保输出目录存在
os.makedirs(output_dir, exist_ok=True)

# 加载数据
print("加载数据...")
# MIR100HG表达水平数据
mir_exp = pd.read_csv(os.path.join(input_dir, 'STAD_Model_MIR100HG_Expression_Levels.csv'))
# 基因表达数据
gene_exp = pd.read_csv(os.path.join(input_dir, 'STAD_Model_Gene_Expression_Features_Test.csv'))
# 甲基化数据
methylation = pd.read_csv(os.path.join(input_dir, 'STAD_Model_Methylation_Features_Test.csv'))

print("MIR100HG表达数据形状:", mir_exp.shape)
print("基因表达数据形状:", gene_exp.shape)
print("甲基化数据形状:", methylation.shape)

# 数据预处理
print("\n数据预处理...")

# 提取标签
y_data = mir_exp[['Sample_ID', 'Group']]
y_data['Group'] = y_data['Group'].map({'High': 1, 'Low': 0})

# 处理基因表达数据
print("\n处理基因表达数据...")
gene_exp_pivot = gene_exp.copy()

# 检查是否存在MIR100HG基因，并将其从特征中删除
if 'MIR100HG' in gene_exp_pivot.index:
    print("检测到MIR100HG基因，将其从特征中删除...")
    gene_exp_pivot = gene_exp_pivot.drop('MIR100HG', axis=0)

gene_exp_pivot = gene_exp_pivot.set_index('HGNC_Symbol')
gene_exp_pivot = gene_exp_pivot.transpose()
gene_exp_pivot = gene_exp_pivot.reset_index()
gene_exp_pivot = gene_exp_pivot.rename(columns={'index': 'Sample_ID'})

# 处理甲基化数据
print("\n处理甲基化数据...")
methylation_pivot = methylation.copy()
methylation_pivot = methylation_pivot.set_index('Probe_ID')
methylation_pivot = methylation_pivot.transpose()
methylation_pivot = methylation_pivot.reset_index()
methylation_pivot = methylation_pivot.rename(columns={'index': 'Sample_ID'})

# 合并数据
print("\n合并数据...")
# 先合并MIR和基因表达
tmp_merged = y_data.merge(gene_exp_pivot, on='Sample_ID', how='inner')
print(f"MIR和基因表达合并后的形状: {tmp_merged.shape}")

# 然后合并甲基化数据
merged_data = tmp_merged.merge(methylation_pivot, on='Sample_ID', how='inner')
print(f"加入甲基化数据后的形状: {merged_data.shape}")

# 检查合并后的数据
if merged_data.shape[0] == 0:
    print("警告: 合并后的数据为空! 检查Sample_ID是否匹配。")
    import sys
    sys.exit(1)

print(f"合并后的数据形状: {merged_data.shape}")
print(f"样本数量: {merged_data.shape[0]}")
print(f"特征数量 (包括Sample_ID和Group): {merged_data.shape[1]}")

# 分割数据为X和y
X = merged_data.drop(['Sample_ID', 'Group'], axis=1)
y = merged_data['Group']
sample_ids = merged_data['Sample_ID']

# 为特征添加前缀
gene_features = gene_exp_pivot.columns.tolist()[1:]  # 跳过Sample_ID
methylation_features = methylation_pivot.columns.tolist()[1:]  # 跳过Sample_ID

gene_prefix = {feature: f"Gene_{feature}" for feature in gene_features}
methylation_prefix = {feature: f"Methylation_{feature}" for feature in methylation_features}

# 重命名列
X = X.rename(columns={**gene_prefix, **methylation_prefix})

print(f"\n添加前缀后的特征数量: {X.shape[1]}")
print(f"基因表达特征数量: {len(gene_features)}")
print(f"甲基化特征数量: {len(methylation_features)}")

# 确保MIR100HG相关特征被删除
mir100hg_cols = [col for col in X.columns if 'MIR100HG' in col]
if len(mir100hg_cols) > 0:
    print(f"删除以下MIR100HG相关特征: {mir100hg_cols}")
    X = X.drop(mir100hg_cols, axis=1)

# 获取基因表达和甲基化特征的列
gene_columns = [col for col in X.columns if col.startswith('Gene_')]
methylation_columns = [col for col in X.columns if col.startswith('Methylation_')]

# 划分训练集和测试集
print("\n划分训练集和测试集...")
X_train, X_test, y_train, y_test, ids_train, ids_test = train_test_split(
    X, y, sample_ids, test_size=0.2, random_state=42, stratify=y
)

print(f"训练集样本数: {X_train.shape[0]}")
print(f"测试集样本数: {X_test.shape[0]}")

# 特征选择函数 - 分别为基因表达和甲基化数据进行特征选择
def select_features(X_train, y_train, var_threshold=0.005, k_gene=200, k_meth=100):
    # 分离基因表达和甲基化特征
    X_train_gene = X_train[gene_columns]
    X_train_meth = X_train[methylation_columns]
    
    print("特征选择前:")
    print(f"总特征数量: {X_train.shape[1]}")
    print(f"基因表达特征数量: {X_train_gene.shape[1]}")
    print(f"甲基化特征数量: {X_train_meth.shape[1]}")
    
    # 1. 对基因表达特征进行方差过滤
    selector_var_gene = VarianceThreshold(threshold=var_threshold)
    X_train_gene_var = selector_var_gene.fit_transform(X_train_gene)
    gene_var_features = X_train_gene.columns[selector_var_gene.get_support()].tolist()
    print(f"方差过滤后的基因表达特征数量: {len(gene_var_features)}")
    
    # 2. 对甲基化特征进行方差过滤
    selector_var_meth = VarianceThreshold(threshold=var_threshold)
    X_train_meth_var = selector_var_meth.fit_transform(X_train_meth)
    meth_var_features = X_train_meth.columns[selector_var_meth.get_support()].tolist()
    print(f"方差过滤后的甲基化特征数量: {len(meth_var_features)}")
    
    # 3. 对基因表达特征进行ANOVA F检验
    X_train_gene_var_df = pd.DataFrame(X_train_gene_var, columns=gene_var_features, index=X_train.index)
    k_gene_actual = min(k_gene, X_train_gene_var_df.shape[1])
    selector_f_gene = SelectKBest(f_classif, k=k_gene_actual)
    _ = selector_f_gene.fit_transform(X_train_gene_var_df, y_train)
    gene_selected_features = X_train_gene_var_df.columns[selector_f_gene.get_support()].tolist()
    print(f"ANOVA选择后的基因表达特征数量: {len(gene_selected_features)}")
    
    # 4. 对甲基化特征进行ANOVA F检验
    X_train_meth_var_df = pd.DataFrame(X_train_meth_var, columns=meth_var_features, index=X_train.index)
    k_meth_actual = min(k_meth, X_train_meth_var_df.shape[1])
    selector_f_meth = SelectKBest(f_classif, k=k_meth_actual)
    _ = selector_f_meth.fit_transform(X_train_meth_var_df, y_train)
    meth_selected_features = X_train_meth_var_df.columns[selector_f_meth.get_support()].tolist()
    print(f"ANOVA选择后的甲基化特征数量: {len(meth_selected_features)}")
    
    # 合并所选特征
    all_selected_features = gene_selected_features + meth_selected_features
    print(f"合并后的所选特征总数: {len(all_selected_features)}")
    print(f"基因表达特征: {len(gene_selected_features)}")
    print(f"甲基化特征: {len(meth_selected_features)}")
    
    # 创建选择矩阵
    combined_support = np.zeros(X_train.shape[1], dtype=bool)
    for i, feature in enumerate(X_train.columns):
        if feature in all_selected_features:
            combined_support[i] = True
    
    return combined_support, all_selected_features

# 使用训练集进行特征选择
print("\n在训练集上进行特征选择...")
feature_support, selected_features = select_features(X_train, y_train)

# 应用特征选择
X_train_selected = X_train.loc[:, feature_support]
X_test_selected = X_test.loc[:, feature_support]

print(f"选择后的特征数量: {X_train_selected.shape[1]}")
print("选择的基因表达特征数量:", len([f for f in X_train_selected.columns if f.startswith('Gene_')]))
print("选择的甲基化特征数量:", len([f for f in X_train_selected.columns if f.startswith('Methylation_')]))

# 特征重要性分析
def feature_importance_analysis(model, X, feature_names):
    # 获取特征重要性
    importances = model.feature_importances_
    
    # 创建一个DataFrame来存储特征重要性
    feature_importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances
    })
    
    # 按重要性排序
    feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)
    
    # 计算百分比
    feature_importance_df['Percentage'] = feature_importance_df['Importance'] / feature_importance_df['Importance'].sum() * 100
    
    return feature_importance_df

# 训练随机森林模型
print("\n训练随机森林模型...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train_selected, y_train)

# 分析特征重要性
print("\n特征重要性分析...")
feature_importance = feature_importance_analysis(rf_model, X_train_selected, X_train_selected.columns)

# 分类特征并计算每类特征的总重要性
gene_importance = feature_importance[feature_importance['Feature'].str.startswith('Gene_')]
methylation_importance = feature_importance[feature_importance['Feature'].str.startswith('Methylation_')]

gene_importance_sum = gene_importance['Importance'].sum()
methylation_importance_sum = methylation_importance['Importance'].sum()
total_importance = gene_importance_sum + methylation_importance_sum

# 打印特征重要性统计
print("\n特征重要性百分比:")
print(f"基因表达特征: {gene_importance_sum / total_importance * 100:.2f}%")
print(f"甲基化特征: {methylation_importance_sum / total_importance * 100:.2f}%")

# 打印按类别分组的Top20特征
print("\n基因表达Top20特征:")
print(gene_importance.head(20).to_string(index=False))

print("\n甲基化Top20特征:")
print(methylation_importance.head(20).to_string(index=False))

print("\n总体Top20特征:")
print(feature_importance.head(20).to_string(index=False))


# 在测试集上评估模型
print("\n在测试集上评估模型...")
y_pred = rf_model.predict(X_test_selected)
y_prob = rf_model.predict_proba(X_test_selected)[:, 1]

# 计算指标
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)
cm = confusion_matrix(y_test, y_pred)

# 打印评估结果
print("\n测试集性能指标:")
print(f"准确率 (Accuracy): {accuracy:.4f}")
print(f"精确率 (Precision): {precision:.4f}")
print(f"召回率 (Recall): {recall:.4f}")
print(f"F1分数: {f1:.4f}")
print(f"AUC: {auc:.4f}")
print("混淆矩阵:")
print(cm)

# 保存模型性能指标
with open(os.path.join(output_dir, 'SKCM_RandomForest_Model_Performance.txt'), 'w') as f:
    f.write(f"测试集性能指标:\n")
    f.write(f"准确率 (Accuracy): {accuracy:.4f}\n")
    f.write(f"精确率 (Precision): {precision:.4f}\n")
    f.write(f"召回率 (Recall): {recall:.4f}\n")
    f.write(f"F1分数: {f1:.4f}\n")
    f.write(f"AUC: {auc:.4f}\n")
    f.write("混淆矩阵:\n")
    f.write(f"{cm[0][0]} {cm[0][1]}\n")
    f.write(f"{cm[1][0]} {cm[1][1]}\n")

# 交叉验证
print("\n执行交叉验证...")
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results = {
    'fold': [],
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1': [],
    'auc': []
}

fold = 1
for train_idx, val_idx in skf.split(X, y):
    print(f"\n执行第{fold}折交叉验证...")
    
    # 获取当前fold的训练和验证数据
    X_fold_train, X_fold_val = X.iloc[train_idx], X.iloc[val_idx]
    y_fold_train, y_fold_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # 在fold内进行特征选择
    print(f"Fold {fold} - 特征选择...")
    fold_feature_support, fold_selected_features = select_features(X_fold_train, y_fold_train)
    
    # 应用特征选择
    X_fold_train_selected = X_fold_train.loc[:, fold_feature_support]
    X_fold_val_selected = X_fold_val.loc[:, fold_feature_support]
    
    # 训练模型
    print(f"Fold {fold} - 训练模型...")
    fold_model = RandomForestClassifier(
        n_estimators=100,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        random_state=42,
        n_jobs=-1
    )
    fold_model.fit(X_fold_train_selected, y_fold_train)
    
    # 预测
    y_fold_pred = fold_model.predict(X_fold_val_selected)
    y_fold_prob = fold_model.predict_proba(X_fold_val_selected)[:, 1]
    
    # 计算性能指标
    fold_accuracy = accuracy_score(y_fold_val, y_fold_pred)
    fold_precision = precision_score(y_fold_val, y_fold_pred)
    fold_recall = recall_score(y_fold_val, y_fold_pred)
    fold_f1 = f1_score(y_fold_val, y_fold_pred)
    fold_auc = roc_auc_score(y_fold_val, y_fold_prob)
    
    # 打印本折结果
    print(f"Fold {fold} - 性能指标:")
    print(f"  准确率: {fold_accuracy:.4f}")
    print(f"  精确率: {fold_precision:.4f}")
    print(f"  召回率: {fold_recall:.4f}")
    print(f"  F1分数: {fold_f1:.4f}")
    print(f"  AUC: {fold_auc:.4f}")
    
    # 保存结果
    cv_results['fold'].append(fold)
    cv_results['accuracy'].append(fold_accuracy)
    cv_results['precision'].append(fold_precision)
    cv_results['recall'].append(fold_recall)
    cv_results['f1'].append(fold_f1)
    cv_results['auc'].append(fold_auc)
    
    # 分析特征重要性
    fold_importance = feature_importance_analysis(fold_model, X_fold_train_selected, X_fold_train_selected.columns)
    
    # 保存本折的特征重要性
    fold_importance.to_csv(os.path.join(output_dir, f'SKCM_RandomForest_Fold{fold}_Feature_Importance.csv'), index=False)
    
    fold += 1

# 计算交叉验证的平均性能
cv_results_df = pd.DataFrame(cv_results)
cv_mean = cv_results_df.mean()

print("\n交叉验证平均性能:")
print(f"平均准确率: {cv_mean['accuracy']:.4f}")
print(f"平均精确率: {cv_mean['precision']:.4f}")
print(f"平均召回率: {cv_mean['recall']:.4f}")
print(f"平均F1分数: {cv_mean['f1']:.4f}")
print(f"平均AUC: {cv_mean['auc']:.4f}")
```
