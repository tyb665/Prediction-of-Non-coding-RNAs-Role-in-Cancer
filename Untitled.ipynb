```python
import matplotlib.pyplot as plt

your_labels = [
    'Lit Review + References (6%)',
    'TF Analysis (8%)',
    'Survival Analysis (5%)',
    'TF Pathway Comparison (3%)',
    'Future Work & Suggestions (3%)',
    'Final Editing (2%)',
    'Methodology & Model Support (11%)'
]

your_sizes = [6, 8, 5, 3, 3, 2, 11]  # 总和为38%

# 外圈数据：你 vs Others
group_labels = ['Me', 'Others']
group_sizes = [38, 62]
group_colors = ['#4caf50', '#cfd8dc']

# 创建图形
fig, ax = plt.subplots(figsize=(9, 9))

# 外圈：团队贡献分布
wedges_outer, _, _ = ax.pie(
    group_sizes,
    radius=1.3,
    labels=group_labels,
    colors=group_colors,
    autopct='%1.1f%%',
    textprops={'fontsize': 10},
    wedgeprops=dict(width=0.3, edgecolor='w')
)

# 内圈：你个人任务细分（带比例说明的标签）
wedges_inner, _ = ax.pie(
    your_sizes,
    radius=1.0,
    labels=your_labels,
    colors=your_colors,
    labeldistance=0.7,
    textprops={'fontsize': 9},
    wedgeprops=dict(width=0.3, edgecolor='w')
)

# 图例
ax.legend(
    wedges_outer,
    group_labels,
    title="Team-Level Contribution",
    loc="center left",
    bbox_to_anchor=(1.0, 0.5),
    fontsize=9
)

ax.set(aspect="equal", title='My Work Contribution and Breakdown in the Group')
plt.tight_layout()
plt.show()
```
