````python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# DEG Analysis Script (Updated version with matplotlib style fix)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multitest import multipletests
import os

# =================== Configuration ===================
expr_file = '/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_TOIL_RSEM_TPM_Levels.csv'
meta_file = '/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_MIR100HG_Expression_Levels.csv'
output_table_dir = '/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED'
output_figure_dir = '/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Figure'
log2fc_threshold = 1.0
padj_threshold = 0.05
use_nonparametric = True
data_is_log2 = True
pseudo_count = 0.001

# åˆ›å»ºç›®å½•
os.makedirs(output_table_dir, exist_ok=True)
os.makedirs(output_figure_dir, exist_ok=True)

# =================== Style Settings ===================
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.2)
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.unicode_minus'] = False

# =================== Load Data ===================
print("Loading data...")
expr_df = pd.read_csv(expr_file)
gene_info = expr_df[['Ensembl_ID', 'HGNC_Symbol']].copy()
expr_values = expr_df.iloc[:, 2:].copy()
expr_values = expr_values.set_index(expr_df['Ensembl_ID'])

meta_df = pd.read_csv(meta_file)
sample_groups = meta_df[['Patient_ID', 'Group']].dropna()
sample_groups['Group'] = sample_groups['Group'].str.strip()
sample_groups = sample_groups.set_index('Patient_ID')

common_samples = expr_values.columns.intersection(sample_groups.index)
expr_values = expr_values[common_samples]
sample_groups = sample_groups.loc[common_samples]
high_samples = sample_groups[sample_groups['Group'] == 'High'].index
low_samples = sample_groups[sample_groups['Group'] == 'Low'].index

print(f"Total genes: {len(expr_values)}, samples: {len(common_samples)}")

# =================== DEG Analysis ===================
print("Performing DEG analysis...")
results = []

for gene in expr_values.index:
    high_vals = expr_values.loc[gene, high_samples].dropna().values
    low_vals = expr_values.loc[gene, low_samples].dropna().values

    if len(high_vals) < 3 or len(low_vals) < 3:
        continue

    high_mean = np.mean(high_vals)
    low_mean = np.mean(low_vals)
    log2fc = high_mean - low_mean if data_is_log2 else np.log2((high_mean + pseudo_count) / (low_mean + pseudo_count))

    if use_nonparametric:
        try:
            stat, pval = stats.mannwhitneyu(high_vals, low_vals, alternative='two-sided')
        except ValueError:
            pval = 1.0
    else:
        stat, pval = stats.ttest_ind(high_vals, low_vals, equal_var=False)

    gene_symbol = gene_info.loc[gene_info['Ensembl_ID'] == gene, 'HGNC_Symbol'].values[0] if gene in gene_info['Ensembl_ID'].values else gene
    results.append({
        'Ensembl_ID': gene,
        'HGNC_Symbol': gene_symbol,
        'Mean_High': high_mean,
        'Mean_Low': low_mean,
        'Log2_Fold_Change': log2fc,
        'P_Value': pval
    })

deg_df = pd.DataFrame(results)
deg_df['Adj_P_Value'] = multipletests(deg_df['P_Value'], method='fdr_bh')[1]
deg_df['Significant'] = (deg_df['Adj_P_Value'] < padj_threshold) & (abs(deg_df['Log2_Fold_Change']) > log2fc_threshold)
deg_df['Regulation'] = deg_df.apply(lambda x: 'Up' if x['Significant'] and x['Log2_Fold_Change'] > 0 else ('Down' if x['Significant'] and x['Log2_Fold_Change'] < 0 else 'Not Significant'), axis=1)

# =================== Save Results ===================
all_results_path = os.path.join(output_table_dir, 'SKCM_DEG_All_Results.csv')
sig_results_path = os.path.join(output_table_dir, 'SKCM_DEG_Significant_Results.csv')
deg_df.to_csv(all_results_path, index=False)
deg_df[deg_df['Significant']].to_csv(sig_results_path, index=False)
print(f"ä¿å­˜å…¨éƒ¨ç»“æœåˆ°: {all_results_path}")
print(f"ä¿å­˜æ˜¾è‘—ç»“æœåˆ°: {sig_results_path}")


# =================== Volcano Plot ===================
print("Plotting volcano plot...")
plt.figure(figsize=(10, 8))
plt.scatter(deg_df['Log2_Fold_Change'], -np.log10(deg_df['P_Value']), c='gray', alpha=0.5)
plt.scatter(deg_df[deg_df['Regulation'] == 'Up']['Log2_Fold_Change'], -np.log10(deg_df[deg_df['Regulation'] == 'Up']['P_Value']), c='red', label='Up')
plt.scatter(deg_df[deg_df['Regulation'] == 'Down']['Log2_Fold_Change'], -np.log10(deg_df[deg_df['Regulation'] == 'Down']['P_Value']), c='blue', label='Down')
plt.axhline(-np.log10(padj_threshold), color='black', linestyle='--')
plt.axvline(log2fc_threshold, color='black', linestyle='--')
plt.axvline(-log2fc_threshold, color='black', linestyle='--')
plt.xlabel('Log2 Fold Change')
plt.ylabel('-Log10 p-value')
plt.title('Volcano Plot: SKCM MIR100HG High vs Low')
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(output_figure_dir, 'SKCM_DEG_Volcano.png'), dpi=300)
plt.close()

print("SKCM DEG analysis completed. Results and plots saved.")
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multitest import multipletests
import os

# è®¾ç½®æ–‡ä»¶è·¯å¾„
methylation_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_Methylation_Levels.csv"
expression_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_MIR100HG_Expression_Levels.csv"
probe_map_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA/probeMap_illuminaMethyl450_hg19_GPL16304_TCGAlegacy"

# 1. è¯»å–æ•°æ®
print("è¯»å–ç”²åŸºåŒ–æ•°æ®...")
meth_data = pd.read_csv(methylation_file, index_col=0)
print(f"ç”²åŸºåŒ–æ•°æ®å½¢çŠ¶: {meth_data.shape}")

print("è¯»å–è¡¨è¾¾æ•°æ®...")
expr_data = pd.read_csv(expression_file)
print(f"è¡¨è¾¾æ•°æ®å½¢çŠ¶: {expr_data.shape}")

# 2. è¯»å–æ¢é’ˆæ˜ å°„æ–‡ä»¶
print("è¯»å–æ¢é’ˆæ˜ å°„æ•°æ®...")
# æ³¨æ„ï¼šæ–‡ä»¶çš„å‰ä¸¤åˆ—æ˜¯#idå’Œgeneï¼Œå¯¹åº”çš„æ˜¯Probe_IDå’ŒHGNC_Symbol
probe_map = pd.read_csv(probe_map_file, sep='\t', comment=None)
# å¦‚æœç¬¬ä¸€åˆ—åç§°ä¸º#idï¼Œåˆ™é‡å‘½åä¸ºProbe_ID
if probe_map.columns[0] == '#id':
    probe_map = probe_map.rename(columns={'#id': 'Probe_ID', 'gene': 'HGNC_Symbol'})
print(f"æ¢é’ˆæ˜ å°„æ•°æ®å½¢çŠ¶: {probe_map.shape}")
print(f"æ¢é’ˆæ˜ å°„æ–‡ä»¶çš„åˆ—å: {probe_map.columns.tolist()}")

# 3. æ•°æ®é¢„å¤„ç†

# è½¬æ¢æ ·æœ¬IDæ ¼å¼ä»¥åŒ¹é…ä¸¤ä¸ªæ•°æ®é›†
def convert_sample_id(sample_id):
    # ä»TCGA-XX-XXXX-XXæ ¼å¼è½¬æ¢ä¸ºTCGA_XX_XXXXæ ¼å¼
    if isinstance(sample_id, str) and sample_id.startswith('TCGA-'):
        parts = sample_id.split('-')
        if len(parts) >= 3:
            return f"TCGA_{parts[1]}_{parts[2]}"
    return sample_id

# åº”ç”¨åˆ°ç”²åŸºåŒ–æ•°æ®çš„åˆ—å
meth_cols = meth_data.columns.tolist()
meth_cols_converted = [convert_sample_id(col) for col in meth_cols]
meth_data.columns = meth_cols_converted

# 4. æ ¹æ®MIR100HGçš„è¡¨è¾¾æ°´å¹³åˆ†ç»„
print("æ ¹æ®MIR100HGè¡¨è¾¾åˆ†ç»„...")
high_group = expr_data[expr_data['Group'] == 'High']['Patient_ID'].tolist()
low_group = expr_data[expr_data['Group'] == 'Low']['Patient_ID'].tolist()

print(f"é«˜è¡¨è¾¾ç»„: {len(high_group)}ä¸ªæ ·æœ¬")
print(f"ä½è¡¨è¾¾ç»„: {len(low_group)}ä¸ªæ ·æœ¬")

# 5. ç­›é€‰å…±æœ‰çš„æ ·æœ¬
high_samples = [s for s in high_group if s in meth_data.columns]
low_samples = [s for s in low_group if s in meth_data.columns]

print(f"å…±æœ‰æ ·æœ¬ - é«˜è¡¨è¾¾ç»„: {len(high_samples)}ä¸ª")
print(f"å…±æœ‰æ ·æœ¬ - ä½è¡¨è¾¾ç»„: {len(low_samples)}ä¸ª")

# 6. æ‰§è¡Œå·®å¼‚ç”²åŸºåŒ–åˆ†æ
print("æ‰§è¡Œå·®å¼‚ç”²åŸºåŒ–åˆ†æ...")

results = []
for probe in meth_data.index:
    # æå–ä¸¤ç»„çš„ç”²åŸºåŒ–å€¼
    high_values = meth_data.loc[probe, high_samples].values
    low_values = meth_data.loc[probe, low_samples].values
    
    # ç§»é™¤NaNå€¼
    high_values = high_values[~np.isnan(high_values)]
    low_values = low_values[~np.isnan(low_values)]
    
    # åªæœ‰å½“ä¸¤ç»„éƒ½æœ‰è¶³å¤Ÿçš„éNaNå€¼æ—¶æ‰è¿›è¡Œtæ£€éªŒ
    if len(high_values) >= 3 and len(low_values) >= 3:
        # è®¡ç®—å¹³å‡å€¼å·®å¼‚
        mean_diff = np.mean(high_values) - np.mean(low_values)
        
        # æ‰§è¡Œtæ£€éªŒ
        t_stat, p_value = stats.ttest_ind(high_values, low_values, equal_var=False)
        
        results.append({
            'Probe_ID': probe,
            'Mean_Diff': mean_diff,
            'p_value': p_value
        })

# è½¬æ¢ä¸ºDataFrame
result_df = pd.DataFrame(results)

# 7. å¤šé‡æ£€éªŒæ ¡æ­£ (FDR)
print("æ‰§è¡Œå¤šé‡æ£€éªŒæ ¡æ­£...")
result_df['adj_p_value'] = multipletests(result_df['p_value'], method='fdr_bh')[1]

# 8. è®¾ç½®æ˜¾è‘—æ€§é˜ˆå€¼
padj_threshold = 0.05
abs_mean_diff_threshold = 0.2

# ç­›é€‰æ˜¾è‘—å·®å¼‚çš„ç”²åŸºåŒ–ä½ç‚¹
sig_dmps = result_df[(result_df['adj_p_value'] < padj_threshold) & 
                     (abs(result_df['Mean_Diff']) > abs_mean_diff_threshold)]

print(f"æ˜¾è‘—å·®å¼‚çš„ç”²åŸºåŒ–ä½ç‚¹: {len(sig_dmps)}")

# 9. ä¸åŸºå› æ³¨é‡Šå¯¹åº”
print("å°†æ¢é’ˆæ˜ å°„åˆ°åŸºå› ...")
# åˆ›å»ºæ¢é’ˆIDåˆ°åŸºå› åçš„æ˜ å°„
# ä½¿ç”¨æ­£ç¡®çš„åˆ—åProbe_IDå’ŒHGNC_Symbol
probe_to_gene = dict(zip(probe_map['Probe_ID'], probe_map['HGNC_Symbol']))

# æ·»åŠ åŸºå› æ³¨é‡Š
sig_dmps['Gene'] = sig_dmps['Probe_ID'].map(lambda x: probe_to_gene.get(x, 'Unknown'))
print(f"æˆåŠŸæ˜ å°„åˆ°åŸºå› çš„æ¢é’ˆæ•°é‡: {sig_dmps['Gene'].value_counts().shape[0] - (1 if 'Unknown' in sig_dmps['Gene'].values else 0)}")

# 10. åˆ›å»ºè¾“å‡ºç›®å½•å¹¶ä¿å­˜ç»“æœ
output_table_dir = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED"
os.makedirs(output_table_dir, exist_ok=True)

# ä¿å­˜å…¨éƒ¨ç»“æœæ–‡ä»¶
all_results_file = os.path.join(output_table_dir, "SKCM_DMA_All_Results.csv")
result_df.to_csv(all_results_file, index=False)

# ä¿å­˜æ˜¾è‘—å·®å¼‚ç»“æœæ–‡ä»¶
sig_results_file = os.path.join(output_table_dir, "SKCM_DMA_Significant_Results.csv")
sig_dmps.sort_values('adj_p_value').to_csv(sig_results_file, index=False)

print(f"å…¨éƒ¨ç»“æœå·²ä¿å­˜ä¸º: {all_results_file}")
print(f"æ˜¾è‘—å·®å¼‚ç»“æœå·²ä¿å­˜ä¸º: {sig_results_file}")
# 11. ç”Ÿæˆçƒ­å›¾å¯è§†åŒ–
print("ç”Ÿæˆçƒ­å›¾å¯è§†åŒ–...")

output_fig_dir = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Figure"
os.makedirs(output_fig_dir, exist_ok=True)

# é€‰æ‹©å‰30ä¸ªæ˜¾è‘—çš„ç”²åŸºåŒ–ä½ç‚¹
top_probes = sig_dmps.sort_values('adj_p_value').head(30)['Probe_ID'].tolist()

if top_probes:
    heatmap_data = meth_data.loc[top_probes, high_samples + low_samples]
    sample_groups = pd.Series(['High'] * len(high_samples) + ['Low'] * len(low_samples),
                             index=high_samples + low_samples)
    
    plt.figure(figsize=(12, 10))
    row_annotations = pd.DataFrame({
        'Gene': [probe_to_gene.get(probe, 'Unknown') for probe in top_probes]
    }, index=top_probes)
    
    col_annotations = pd.DataFrame({
        'Group': sample_groups
    })
    
    cmap = sns.diverging_palette(220, 20, as_cmap=True)
    
    g = sns.clustermap(heatmap_data, 
                      cmap=cmap,
                      z_score=0,
                      row_colors=row_annotations['Gene'].astype('category').cat.codes.map({i: plt.cm.Set3(i) for i in range(20)}),
                      col_colors=col_annotations['Group'].map({'High': 'red', 'Low': 'blue'}),
                      figsize=(12, 10),
                      dendrogram_ratio=(.1, .2),
                      cbar_pos=(.02, .32, .03, .2))
    
    heatmap_file = os.path.join(output_fig_dir, "DMA_Heatmap.png")
    plt.savefig(heatmap_file, dpi=300, bbox_inches='tight')
    print(f"çƒ­å›¾å·²ä¿å­˜ä¸º: {heatmap_file}")
else:
    print("æ²¡æœ‰è¶³å¤Ÿçš„æ˜¾è‘—å·®å¼‚ç”²åŸºåŒ–ä½ç‚¹ç”¨äºçƒ­å›¾")

# 11. ç”Ÿæˆçƒ­å›¾å¯è§†åŒ–
print("ç”Ÿæˆçƒ­å›¾å¯è§†åŒ–...")

output_fig_dir = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Figure"
os.makedirs(output_fig_dir, exist_ok=True)

# é€‰æ‹©å‰30ä¸ªæ˜¾è‘—çš„ç”²åŸºåŒ–ä½ç‚¹
top_probes = sig_dmps.sort_values('adj_p_value').head(30)['Probe_ID'].tolist()

if top_probes:
    heatmap_data = meth_data.loc[top_probes, high_samples + low_samples]
    sample_groups = pd.Series(['High'] * len(high_samples) + ['Low'] * len(low_samples),
                             index=high_samples + low_samples)
    
    plt.figure(figsize=(12, 10))
    row_annotations = pd.DataFrame({
        'Gene': [probe_to_gene.get(probe, 'Unknown') for probe in top_probes]
    }, index=top_probes)
    
    col_annotations = pd.DataFrame({
        'Group': sample_groups
    })
    
    cmap = sns.diverging_palette(220, 20, as_cmap=True)
    
    g = sns.clustermap(heatmap_data, 
                      cmap=cmap,
                      z_score=0,
                      row_colors=row_annotations['Gene'].astype('category').cat.codes.map({i: plt.cm.Set3(i) for i in range(20)}),
                      col_colors=col_annotations['Group'].map({'High': 'red', 'Low': 'blue'}),
                      figsize=(12, 10),
                      dendrogram_ratio=(.1, .2),
                      cbar_pos=(.02, .32, .03, .2))
    
    heatmap_file = os.path.join(output_fig_dir, "SKCM_DMA_Heatmap.png")
    plt.savefig(heatmap_file, dpi=300, bbox_inches='tight')
    print(f"çƒ­å›¾å·²ä¿å­˜ä¸º: {heatmap_file}")
else:
    print("æ²¡æœ‰è¶³å¤Ÿçš„æ˜¾è‘—å·®å¼‚ç”²åŸºåŒ–ä½ç‚¹ç”¨äºçƒ­å›¾")
# 13. å¯Œé›†åˆ†æè¾“å‡ºå›¾ç‰‡è·¯å¾„
...
enrichment_plot_file = os.path.join(output_fig_dir, "SKCM_DMA_Enrichment.png")
plt.savefig(enrichment_plot_file, dpi=300, bbox_inches='tight')
print(f"å¯Œé›†åˆ†æå›¾å·²ä¿å­˜ä¸º: {enrichment_plot_file}")
````

````python
import pandas as pd
import mygene
import os
import re

# === Step 0: è®¾ç½®è·¯å¾„ ===
input_path = '/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA/geneAnnotation_hg19_basicgenes.txt'
output_dir = '/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/'
output_path = os.path.join(output_dir, 'Promoter_Region.csv')

# åˆ›å»ºè¾“å‡ºç›®å½•ï¼ˆå¦‚æœä¸å­˜åœ¨ï¼‰
os.makedirs(output_dir, exist_ok=True)

# === Step 1: è¯»å–å’Œæ¸…æ´—åˆ— ===
df = pd.read_csv(input_path, sep='\t', dtype=str)

# é‡å‘½ååˆ—
df = df.rename(columns={
    'chr': 'Promoter_Chr',
    'start': 'Promoter_Start',
    'end': 'Promoter_End',
    'width': 'Promoter_Width',
    'id': 'Promoter_ID',
    'gene_id': 'UCSC_Entrez_ID'  # âœ… é‡å‘½å
})

# åˆ é™¤ä¸éœ€è¦çš„åˆ—
df = df.drop(columns=['strand', 'tx_id', 'symbol', 'type'], errors='ignore')

# åˆ é™¤ç©ºå€¼
df = df[df['UCSC_Entrez_ID'].notnull() & (df['UCSC_Entrez_ID'].str.strip() != '')]

# === Step 2: åˆ¤æ–­å“ªäº›æ˜¯ UCSC IDï¼Œå“ªäº›æ˜¯ Entrez ID ===
def is_ucsc(x):
    return bool(re.match(r'^uc\d+[a-z]+\d*$', x.strip().split('.')[0]))

df['ID_Type'] = df['UCSC_Entrez_ID'].apply(lambda x: 'ucsc' if is_ucsc(x) else 'entrezgene')
df['query_id'] = df['UCSC_Entrez_ID'].str.replace(r'\.\d+$', '', regex=True)  # å»æ‰ç‰ˆæœ¬å·

# === Step 3: ä½¿ç”¨ mygene æ‰¹é‡æŸ¥è¯¢ ===
mg = mygene.MyGeneInfo()

# åˆ†å¼€æŸ¥ UCSC å’Œ Entrez ID
results = []

for scope in ['ucsc', 'entrezgene']:
    ids_to_query = df[df['ID_Type'] == scope]['query_id'].unique().tolist()
    print(f"ğŸ” æ­£åœ¨æŸ¥è¯¢ {len(ids_to_query)} ä¸ª {scope} ID...")
    res = mg.querymany(ids_to_query, scopes=scope, fields='symbol', species='human')
    results.extend(res)

# è½¬ä¸º DataFrame
query_df = pd.DataFrame(results)

# åªä¿ç•™æˆåŠŸæ˜ å°„çš„è®°å½•
query_df = query_df[query_df['symbol'].notnull() & query_df['query'].notnull()]
query_df = query_df[['query', 'symbol']].drop_duplicates()
query_df.columns = ['query_id', 'HGNC_Symbol']

# === Step 4: åˆå¹¶å›ä¸»è¡¨ ===
df = df.merge(query_df, on='query_id', how='left')
df = df.drop(columns=['ID_Type', 'query_id'])

# === Step 5: ä¿å­˜ ===
df.to_csv(output_path, index=False)
print(f"âœ… å¤„ç†å®Œæˆï¼æ–‡ä»¶å·²ä¿å­˜è‡³ï¼š{output_path}")
````

````python
import pandas as pd
import os

# ======= è®¾ç½®è·¯å¾„ ========
dma_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_DMA_Significant_Results.csv"
probe_map_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Probe_Map.csv"
promoter_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Promoter_Region.csv"
deg_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_DEG_Significant_Results.csv"
output_dir = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED"

os.makedirs(output_dir, exist_ok=True)

# ======= è¯»å–æ•°æ® ========
dma_df = pd.read_csv(dma_file)
probe_df = pd.read_csv(probe_map_file)
promoter_df = pd.read_csv(promoter_file)
deg_df = pd.read_csv(deg_file)

# ======= åˆå¹¶ DMA å’Œ Probe åæ ‡ ========
dma_annotated = dma_df.merge(probe_df, on="Probe_ID", how="left")

# ======= åˆ¤æ–­ Probe æ˜¯å¦è½åœ¨å¯åŠ¨å­åŒºåŸŸ ========
matched = []

for _, row in dma_annotated.iterrows():
    hits = promoter_df[
        (promoter_df['Promoter_Chr'] == row['Probe_Chr']) &
        (promoter_df['Promoter_Start'] <= row['Probe_Chrom_Start']) &
        (promoter_df['Promoter_End'] >= row['Probe_Chrom_End'])
    ]
    for _, hit in hits.iterrows():
        matched.append({
            "Probe_ID": row["Probe_ID"],
            "HGNC_Symbol": hit["HGNC_Symbol"],
            "Mean_Diff": row["Mean_Diff"],
            "adj_p_value": row["adj_p_value"]
        })

# è½¬æ¢ä¸º DataFrame
promoter_hits_df = pd.DataFrame(matched)

# ======= ä¿å­˜æ‰€æœ‰å¯åŠ¨å­å†… Probe ========
promoter_hits_path = os.path.join(output_dir, "SKCM_Methylation_Probes_in_Promoters.csv")
promoter_hits_df.to_csv(promoter_hits_path, index=False)
print(f"âœ… æ‰€æœ‰å¯åŠ¨å­åŒºåŸŸçš„ probe ä¿å­˜è‡³ï¼š{promoter_hits_path}")

# ======= è”åˆ DEG è¡¨è¾¾æ•°æ® ========
deg_filtered = deg_df[["HGNC_Symbol", "Log2_Fold_Change"]]
merged = promoter_hits_df.merge(deg_filtered, on="HGNC_Symbol", how="left")

# ======= å¯åŠ¨å­ç”²åŸºåŒ–â†‘ + è¡¨è¾¾â†“ ç­›é€‰ ========
up_meth_down_expr = merged[
    (merged["Mean_Diff"] > 0.15) & 
    (merged["Log2_Fold_Change"] < -0.5)
].dropna()

up_meth_down_path = os.path.join(output_dir, "SKCM_Promoters_Methylation_Up_Expression_Down.csv")
up_meth_down_expr.to_csv(up_meth_down_path, index=False)
print(f"âœ… å¯åŠ¨å­ç”²åŸºåŒ–â†‘ + è¡¨è¾¾â†“ çš„åŸºå› ä¿å­˜è‡³ï¼š{up_meth_down_path}")

# ======= å¯åŠ¨å­ç”²åŸºåŒ–â†“ + è¡¨è¾¾â†‘ ç­›é€‰ ========
down_meth_up_expr = merged[
    (merged["Mean_Diff"] < -0.15) & 
    (merged["Log2_Fold_Change"] > 0.5)
].dropna()

down_meth_up_path = os.path.join(output_dir, "SKCM_Promoters_Methylation_Down_Expression_Up.csv")
down_meth_up_expr.to_csv(down_meth_up_path, index=False)
print(f"âœ… å¯åŠ¨å­ç”²åŸºåŒ–â†“ + è¡¨è¾¾â†‘ çš„åŸºå› ä¿å­˜è‡³ï¼š{down_meth_up_path}")
````

| åˆ—å          | å«ä¹‰è¯´æ˜                               |
|---------------|----------------------------------------|
| `HGNC_Symbol` | åŸºå› æ ‡å‡†åç§°ï¼ˆå¦‚ TP53ã€BRCA1ï¼‰         |
| `Gene_Chr`    | åŸºå› æ‰€åœ¨çš„æŸ“è‰²ä½“ç¼–å·ï¼ˆå¦‚ chr1, chr17ï¼‰ |
| `Gene_Start`  | åŸºå› åœ¨æŸ“è‰²ä½“ä¸Šçš„èµ·å§‹ä½ç½®ï¼ˆ1-basedï¼‰    |
| `Gene_End`    | åŸºå› åœ¨æŸ“è‰²ä½“ä¸Šçš„ç»“æŸä½ç½®               |

### ğŸ§  ä½¿ç”¨ç›®çš„ï¼š

æˆ‘ä»¬åç»­çš„åˆ†æç›®æ ‡æ˜¯è¯†åˆ«**ä½äºåŸºå› ä½“åŒºåŸŸ**çš„ç”²åŸºåŒ–æ¢é’ˆï¼ˆå³ä¸åœ¨å¯åŠ¨å­åŒºåŸŸã€ä½†åœ¨åŸºå› ç»“æ„èŒƒå›´å†…ï¼‰ï¼Œå¹¶è¿›ä¸€æ­¥è”åˆè¡¨è¾¾æ•°æ®æŒ–æ˜ï¼š

- **åŸºå› ä½“ç”²åŸºåŒ–å‡é«˜ + è¡¨è¾¾ä¸‹è°ƒ**ï¼ˆå¯èƒ½å…·æœ‰è°ƒæ§åŠŸèƒ½ï¼‰ï¼›
- **åŸºå› ä½“ç”²åŸºåŒ–ä¸‹é™ + è¡¨è¾¾ä¸Šè°ƒ**ï¼ˆå¯èƒ½è§£é™¤æ²‰é»˜ï¼‰ï¼›

é€šè¿‡æ„å»º `Gene_Region.csv`ï¼Œæˆ‘ä»¬å¯ä»¥ç²¾ç¡®åˆ¤æ–­æ¯ä¸ªå·®å¼‚ç”²åŸºåŒ– probe æ˜¯å¦å‘½ä¸­æŸä¸ªåŸºå› ä½“åŒºåŸŸï¼Œä»è€Œé¿å…å°†éåŠŸèƒ½åŒºï¼ˆå¦‚ intergenic åŒºåŸŸï¼‰è¯¯è®¤ä¸ºåŸºå› ç›¸å…³è°ƒæ§ä½ç‚¹ã€‚


````python
import pandas as pd
import gzip
import requests
import os

# ========== Step 1: ä¸‹è½½ GTF æ–‡ä»¶ï¼ˆEnsembl GRCh38ï¼‰ ==========
gtf_url = "http://ftp.ensembl.org/pub/release-109/gtf/homo_sapiens/Homo_sapiens.GRCh38.109.gtf.gz"
gtf_path = "Homo_sapiens.GRCh38.109.gtf.gz"

if not os.path.exists(gtf_path):
    print("ğŸ”½ æ­£åœ¨ä¸‹è½½ GTF æ–‡ä»¶...")
    r = requests.get(gtf_url, stream=True)
    with open(gtf_path, "wb") as f:
        for chunk in r.iter_content(chunk_size=8192):
            f.write(chunk)
    print("âœ… ä¸‹è½½å®Œæˆï¼")

# ========== Step 2: æå– gene åŒºåŸŸ ==========
genes = []
with gzip.open(gtf_path, 'rt') as f:
    for line in f:
        if line.startswith("#"):
            continue
        parts = line.strip().split("\t")
        if parts[2] != "gene":
            continue
        chr_, source, type_, start, end, score, strand, phase, attributes = parts
        attr_dict = {k.strip(): v.strip('"') for k, v in [field.strip().split(" ") for field in attributes.strip(";").split(";") if field]}
        if 'gene_name' in attr_dict:
            # æ ‡å‡†åŒ–æŸ“è‰²ä½“æ ¼å¼ä¸º chrN
            chr_fixed = chr_ if chr_.startswith("chr") else f"chr{chr_}"
            gene = {
                "HGNC_Symbol": attr_dict['gene_name'],
                "Gene_Chr": chr_fixed,
                "Gene_Start": int(start),
                "Gene_End": int(end)
            }
            genes.append(gene)

# ========== Step 3: ä¿å­˜ä¸ºæŒ‡å®šä½ç½®çš„ Gene_Region.csv ==========
output_path = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Gene_Region.csv"
gene_df = pd.DataFrame(genes)
gene_df = gene_df[gene_df["Gene_Chr"].str.startswith("chr")]  # ä¿ç•™æ ‡å‡†æŸ“è‰²ä½“æ ¼å¼
gene_df.to_csv(output_path, index=False)
print(f"ğŸ“ åŸºå› ä½“æ³¨é‡Šæ–‡ä»¶å·²ä¿å­˜è‡³ï¼š{output_path}")
````

- åç»­å¯ç»“åˆé€šè·¯å¯Œé›†åˆ†ææˆ–å¯è§†åŒ–è¿›ä¸€æ­¥æŒ–æ˜è¿™äº›å€™é€‰åŸºå› çš„åŠŸèƒ½èƒŒæ™¯ã€‚



````python
import pandas as pd

# ======= è·¯å¾„é…ç½® =======
dma_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_DMA_Significant_Results.csv"
probe_map_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Probe_Map.csv"
promoter_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Promoter_Region.csv"
deg_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/SKCM_DEG_Significant_Results.csv"
gene_region_file = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED/Gene_Region.csv"
output_dir = "/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED"

# ======= æ•°æ®è¯»å– =======
dma_df = pd.read_csv(dma_file)
probe_df = pd.read_csv(probe_map_file)
promoter_df = pd.read_csv(promoter_file)
deg_df = pd.read_csv(deg_file)
gene_region_df = pd.read_csv(gene_region_file)

# åˆå¹¶ probe åæ ‡ä¿¡æ¯
dma_annotated = dma_df.merge(probe_df, on="Probe_ID", how="left")

# ç­›é™¤å¯åŠ¨å­åŒºåŸŸå†… probe
promoter_probes = set()
for _, row in promoter_df.iterrows():
    hits = dma_annotated[
        (dma_annotated['Probe_Chr'] == row['Promoter_Chr']) &
        (dma_annotated['Probe_Chrom_Start'] >= row['Promoter_Start']) &
        (dma_annotated['Probe_Chrom_End'] <= row['Promoter_End'])
    ]
    promoter_probes.update(hits['Probe_ID'].tolist())

non_promoter_df = dma_annotated[~dma_annotated['Probe_ID'].isin(promoter_probes)]

# åŒ¹é…åŸºå› ä½“åŒºåŸŸ
matched = []
for _, row in non_promoter_df.iterrows():
    hits = gene_region_df[
        (gene_region_df['Gene_Chr'] == row['Probe_Chr']) &
        (gene_region_df['Gene_Start'] <= row['Probe_Chrom_Start']) &
        (gene_region_df['Gene_End'] >= row['Probe_Chrom_End'])
    ]
    for _, hit in hits.iterrows():
        matched.append({
            "Probe_ID": row["Probe_ID"],
            "HGNC_Symbol": hit["HGNC_Symbol"],
            "Mean_Diff": row["Mean_Diff"],
            "adj_p_value": row["adj_p_value"]
        })

gene_body_hits_df = pd.DataFrame(matched)

# è”åˆ DEG è¡¨è¾¾æ•°æ®
deg_filtered = deg_df[["HGNC_Symbol", "Log2_Fold_Change"]]
merged = gene_body_hits_df.merge(deg_filtered, on="HGNC_Symbol", how="left")

# å¯åŠ¨å­å¤–ç”²åŸºåŒ–ä¸‹é™ + è¡¨è¾¾ä¸Šå‡
hypo_up = merged[
    (merged["Mean_Diff"] < -0.15) & 
    (merged["Log2_Fold_Change"] > 0.5)
].dropna()

# å¯åŠ¨å­å¤–ç”²åŸºåŒ–ä¸Šå‡ + è¡¨è¾¾ä¸‹è°ƒ
hyper_down = merged[
    (merged["Mean_Diff"] > 0.15) & 
    (merged["Log2_Fold_Change"] < -0.5)
].dropna()

# ä¿å­˜ç»“æœ
hypo_up.to_csv(f"{output_dir}/SKCM_GeneBody_Methylation_Down_Expression_Up.csv", index=False)
hyper_down.to_csv(f"{output_dir}/SKCM_GeneBody_Methylation_Up_Expression_Down.csv", index=False)
gene_body_hits_df.to_csv(f"{output_dir}/SKCM_Methylation_Probes_in_GeneBody.csv", index=False)

print("âœ… åˆ†æå®Œæˆï¼Œç»“æœå·²ä¿å­˜ï¼š")
print(" - å¯åŠ¨å­å¤–ç”²åŸºåŒ–ä¸‹é™ + è¡¨è¾¾ä¸Šå‡ â†’ SKCM_GeneBody_Methylation_Down_Expression_Up.csv")
print(" - å¯åŠ¨å­å¤–ç”²åŸºåŒ–ä¸Šå‡ + è¡¨è¾¾ä¸‹è°ƒ â†’ SKCM_GeneBody_Methylation_Up_Expression_Down.csv")
print(" - æ‰€æœ‰å‘½ä¸­åŸºå› ä½“åŒºåŸŸçš„ probe â†’ SKCM_Methylation_Probes_in_GeneBody.csv")
````

````python
"""
Author: Weilin He
Description: Reads and analyzes CSV files related to methylation and expression data.
"""

import pandas as pd
import numpy as np
import os

# Directory containing the CSV files
base_dir = '/Users/heweilin/Desktop/M-28-SKCM/NTU_DATA_CLEANED'

# List of CSV files to process
files = [
    'SKCM_GeneBody_Methylation_Down_Expression_Up.csv',
    'SKCM_GeneBody_Methylation_Up_Expression_Down.csv',
    'SKCM_Methylation_Probes_in_GeneBody.csv',
    'SKCM_Methylation_Probes_in_Promoters.csv',
    'SKCM_Promoters_Methylation_Down_Expression_Up.csv',
    'SKCM_Promoters_Methylation_Up_Expression_Down.csv'
]

results = {}

# Function to load and analyze CSV files
def analyze_csv(file_path):
    df = pd.read_csv(file_path)
    analysis = {
        'row_count': len(df),
        'column_names': df.columns.tolist(),
        'sample_data': df.head(3).to_dict(orient='records')
    }

    # Detailed analysis for specific files
    if 'Methylation_Probes_in_GeneBody.csv' in file_path:
        mean_diff = df['Mean_Diff'].dropna()
        analysis['mean_diff_stats'] = {
            'min': mean_diff.min(),
            'max': mean_diff.max(),
            'avg': mean_diff.mean()
        }

    if 'Methylation_Probes_in_Promoters.csv' in file_path:
        mean_diff = df['Mean_Diff'].dropna()
        positive = (mean_diff > 0).sum()
        negative = (mean_diff < 0).sum()
        analysis['mean_diff_stats'] = {
            'min': mean_diff.min(),
            'max': mean_diff.max(),
            'avg': mean_diff.mean(),
            'positive_count': positive,
            'negative_count': negative,
            'positive_percentage': f"{positive / len(mean_diff) * 100:.2f}%",
            'negative_percentage': f"{negative / len(mean_diff) * 100:.2f}%"
        }

    if 'Promoters_Methylation_Down_Expression_Up.csv' in file_path:
        log2fc = df['Log2_Fold_Change'].dropna()
        mean_diff = df['Mean_Diff'].dropna()
        analysis['expression_stats'] = {
            'min_log2fc': log2fc.min(),
            'max_log2fc': log2fc.max(),
            'avg_log2fc': log2fc.mean(),
            'avg_mean_diff': mean_diff.mean()
        }

    return analysis

# Process each file
for file in files:
    file_path = os.path.join(base_dir, file)
    try:
        results[file] = analyze_csv(file_path)
    except Exception as e:
        results[file] = {'error': str(e)}

# Generate summary statistics
summary = {
    'total_promoter_probes': results.get('SKCM_Methylation_Probes_in_Promoters.csv', {}).get('row_count', 0),
    'total_gene_body_probes': results.get('SKCM_Methylation_Probes_in_GeneBody.csv', {}).get('row_count', 0),
    'promoter_methyl_down_expr_up': results.get('SKCM_Promoters_Methylation_Down_Expression_Up.csv', {}).get('row_count', 0),
    'promoter_methyl_up_expr_down': results.get('SKCM_Promoters_Methylation_Up_Expression_Down.csv', {}).get('row_count', 0),
    'gene_body_methyl_down_expr_up': results.get('SKCM_GeneBody_Methylation_Down_Expression_Up.csv', {}).get('row_count', 0),
    'gene_body_methyl_up_expr_down': results.get('SKCM_GeneBody_Methylation_Up_Expression_Down.csv', {}).get('row_count', 0)
}

# Display results
print("Summary of Analysis:", summary)

if 'SKCM_Promoters_Methylation_Down_Expression_Up.csv' in results:
    print("Detailed Expression Stats (Promoters Down Methylation & Expression Up):", 
          results['SKCM_Promoters_Methylation_Down_Expression_Up.csv'].get('expression_stats', {}))

if 'SKCM_GeneBody_Methylation_Down_Expression_Up.csv' in results:
    print("Gene Body Row Count:", results['SKCM_GeneBody_Methylation_Down_Expression_Up.csv']['row_count'])
    print("Gene Body Sample Data:", results['SKCM_GeneBody_Methylation_Down_Expression_Up.csv']['sample_data'])

if 'SKCM_Methylation_Probes_in_Promoters.csv' in results:
    print("Promoter Region Methylation Distribution:", 
          results['SKCM_Methylation_Probes_in_Promoters.csv'].get('mean_diff_stats', {}))

print("Count of Promoter Methylation Up & Expression Down Genes:", 
      summary['promoter_methyl_up_expr_down'])
````
